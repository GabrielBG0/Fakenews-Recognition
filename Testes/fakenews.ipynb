```python
import pickle
import numpy as np
import pandas as pd


documents = pickle.load(
    open("../Datasets/FakeNewsNet/dict_documents.pkl", "rb"))
indexes = np.load('../Datasets/FakeNewsNet/indexes.npy')
lbls = np.load('../Datasets/FakeNewsNet/labels.npy')

texts = []
labels = []
for i in range(len(indexes)):
    index = indexes[i]
    texts.append(documents[index])
    if lbls[i] == np.str_("-1"):
        labels.append(0)
    else:
        labels.append(1)

df = pd.DataFrame(data={"text": texts, "label": labels})
df
```
